```python
# setup
import pandas as pd
# read and import the csv file exported from MS Lists
df = pd.read_csv('assets.csv', header=0,index_col=None)
df.head()
```

```python

# fill down property
propertycol = df['Property']
propertycol.fillna(method='ffill', inplace=True)
# fill down building
buildingcol = df['Building']
buildingcol.fillna(method='ffill', inplace=True)
# fill down floor
floorcol = df['Floor']
floorcol.fillna(method='ffill', inplace=True)

# remove line if Asset Name is empty
df = df.dropna(subset='Asset Name')

#remove last 4 columns (Created By, Modified By, Asset Name Lookup and Created Date/Time)
df = df.iloc[:, :-4]
# remove Images column
df.drop('Image', inplace=True, axis=1)

# preview
#df.head(10)

# convert Quantity column to numeric
df['Quantity'] = pd.to_numeric(df['Quantity'])
# repeat rows x Quantity
df = df.loc[df.index.repeat(df['Quantity'])]

# preview
df.head()
```

```python
# export to Export.xlsx file
df.to_excel('assets_cleaned.xlsx', index=None)
```
